```python
from sklearn.model_selection import train_test_split
import pandas as pd
import os
import time
```

```python
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/MyDrive/Colab Notebooks/Albert_on_Instagram_Comment/data')
#print(f"Current directory: {os.getcwd()}")
```

```python
# Load dataset
df = pd.read_csv('data_komen_preprocessed.csv')
```

```python
# Split the data into train and test sets (90% train, 10% test)
start = time.time()

train_data, test_data = train_test_split(df, test_size=0.1, random_state=42)
train_data.to_csv('data_training_90.csv', index=False)
test_data.to_csv('data_testing_10.csv', index=False)

print("Spliting completed and saved to 'data_training_90.csv' and 'data_testing_10.csv")
elapsed = (time.time() - start) / 60
print(f"Processing time: {elapsed} minutes")
```
