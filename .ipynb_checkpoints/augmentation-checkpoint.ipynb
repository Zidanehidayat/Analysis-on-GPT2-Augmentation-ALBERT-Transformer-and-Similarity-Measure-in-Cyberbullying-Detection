```python
def noRs(n, y):
    start = time.time()

    # Define arguments for the augment function
    input_file = "data/data_training_90.txt"
    output_file = f"data/data_baru/data_training_90_{n}_{y[0]}_NO_RS.txt"
    num_aug = n
    alpha_sr = y[1]
    alpha_rd = y[1]
    alpha_ri = y[1]
    alpha_rs = 0  # Set to 0 as required


    # Call the augment function directly with the arguments
    result = augment.gen_eda(input_file, output_file,
                             alpha_sr, alpha_rd,
                             alpha_ri, alpha_rs, num_aug)
    print(result)

    # Calculate elapsed time
    elapsed = (time.time() - start) / 60

    # Print the result if augment.augment() returns any output
    print(result)
    print(f"Elapsed time: {elapsed} minutes\n")
```

```python
def run_threads():
    alpha = [['a', 0.05]]
    arguments = [1]  # List of arguments for the function
    threads = []

    for arg in arguments:
        for alp in alpha:
            thread = threading.Thread(target=noRs, args=(arg,alp))
            threads.append(thread)
            thread.start()

    # Wait for all threads to complete
    for thread in threads:
        thread.join()

if __name__ == "__main__":
    run_threads()
```
