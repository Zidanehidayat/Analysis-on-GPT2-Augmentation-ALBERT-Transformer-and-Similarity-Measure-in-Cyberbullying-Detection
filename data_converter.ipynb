```python
import pandas as pd
```

```python
# Create DataFrame
df = pd.read_csv("data/smsa_doc-sentiment-prosa/train_preprocess.tsv", sep="\t", header = None)

# Use 'others' column as the main sentiment
df['label'] = df[1].map({"neutral": 0, "positive": 1, "negative": -1})
df['sentence'] = df[0]

# Drop other sentiment columns
df = df[["label", "sentence"]]
# Save and Display the updated DataFrame
df.to_csv('data/smsa_doc-sentiment-prosa/smsa_train_preprocess.txt', sep="\t", header=False ,encoding='utf-16', index=False)
df.head()
```
