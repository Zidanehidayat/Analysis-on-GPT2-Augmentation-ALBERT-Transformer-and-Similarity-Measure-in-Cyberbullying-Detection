```python
from sklearn.model_selection import train_test_split
import pandas as pd
import os
import time
```

```python
# Load dataset
df = pd.read_csv('data/dataset_FND.csv')
```

```python
# Split the data into train and test sets (90% train, 10% test)
start = time.time()

train_data, test_data = train_test_split(df, test_size=0.1, random_state=42)
train_data.to_csv('data/data_FND_training_90.txt', sep="\t", header=0,encoding='utf-16', index=False)
test_data.to_csv('data/data_FND_testing_10.txt',  sep="\t", header=0,encoding='utf-16', index=False)

print("Spliting completed and saved to 'data_training_90.csv' and 'data_testing_10.csv")
elapsed = (time.time() - start) / 60
print(f"Processing time: {elapsed} minutes")
```
